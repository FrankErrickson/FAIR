# Investigating the differences between including and ommitting fossil methane fraction

```python
import fair
from fair.forward import fair_scm
from fair.RCPs import rcp26, rcp45, rcp60, rcp85
import matplotlib.pyplot as pl
import numpy as np
```

```python
C = {}
F = {}
T = {}
```

```python
fossil_ch4 = {
    'rcp26_exc' : 0,
    'rcp45_exc' : 0,
    'rcp60_exc' : 0,
    'rcp85_exc' : 0,
    'rcp26_inc' : rcp26.fossilCH4_frac,
    'rcp45_inc' : rcp45.fossilCH4_frac,
    'rcp60_inc' : rcp60.fossilCH4_frac,
    'rcp85_inc' : rcp85.fossilCH4_frac
}
```

```python
emissions = {
    'rcp26_exc' : rcp26.Emissions.emissions,
    'rcp45_exc' : rcp45.Emissions.emissions,
    'rcp60_exc' : rcp60.Emissions.emissions,
    'rcp85_exc' : rcp85.Emissions.emissions,
    'rcp26_inc' : rcp26.Emissions.emissions,
    'rcp45_inc' : rcp45.Emissions.emissions,
    'rcp60_inc' : rcp60.Emissions.emissions,
    'rcp85_inc' : rcp85.Emissions.emissions
}
```

```python
for scen in emissions.keys():
    C[scen], F[scen], T[scen] = fair_scm(
        emissions=emissions[scen],
        fossilCH4_frac=fossil_ch4[scen]
    )
```

```python
fig, ax = pl.subplots(2,2,figsize=(16,9))
ax[0,0].plot(np.arange(1765, 2501), C['rcp26_exc'][:,0], label='excluding CH4 oxidation')
ax[0,0].plot(np.arange(1765, 2501), C['rcp26_inc'][:,0], label='including CH4 oxidation')
ax[0,1].plot(np.arange(1765, 2501), C['rcp45_exc'][:,0])
ax[0,1].plot(np.arange(1765, 2501), C['rcp45_inc'][:,0])
ax[1,0].plot(np.arange(1765, 2501), C['rcp60_exc'][:,0])
ax[1,0].plot(np.arange(1765, 2501), C['rcp60_inc'][:,0])
ax[1,1].plot(np.arange(1765, 2501), C['rcp85_exc'][:,0])
ax[1,1].plot(np.arange(1765, 2501), C['rcp85_inc'][:,0])
ax[0,0].set_title('RCP2.6 CO2 concentrations')
ax[0,1].set_title('RCP4.5 CO2 concentrations')
ax[1,0].set_title('RCP6.0 CO2 concentrations')
ax[1,1].set_title('RCP8.5 CO2 concentrations')
ax[0,0].legend()
```

```python
fig, ax = pl.subplots(figsize=(16,9))
ax.plot(np.arange(1765, 2501),C['rcp26_inc'][:,0]-C['rcp26_exc'][:,0], label='RCP2.6')
ax.plot(np.arange(1765, 2501),C['rcp45_inc'][:,0]-C['rcp45_exc'][:,0], label='RCP4.5')
ax.plot(np.arange(1765, 2501),C['rcp60_inc'][:,0]-C['rcp60_exc'][:,0], label='RCP6.0')
ax.plot(np.arange(1765, 2501),C['rcp85_inc'][:,0]-C['rcp85_exc'][:,0], label='RCP8.5')
ax.legend()
pl.title('Difference between including and excluding CH4 oxidation to CO2, ppm')
```

```python
fig, ax = pl.subplots(2,2,figsize=(16,9))
ax[0,0].plot(np.arange(1765, 2501), T['rcp26_exc'], label='excluding CH4 oxidation')
ax[0,0].plot(np.arange(1765, 2501), T['rcp26_inc'], label='including CH4 oxidation')
ax[0,1].plot(np.arange(1765, 2501), T['rcp45_exc'])
ax[0,1].plot(np.arange(1765, 2501), T['rcp45_inc'])
ax[1,0].plot(np.arange(1765, 2501), T['rcp60_exc'])
ax[1,0].plot(np.arange(1765, 2501), T['rcp60_inc'])
ax[1,1].plot(np.arange(1765, 2501), T['rcp85_exc'])
ax[1,1].plot(np.arange(1765, 2501), T['rcp85_inc'])
ax[0,0].set_title('RCP2.6 temperature')
ax[0,1].set_title('RCP4.5 temperature')
ax[1,0].set_title('RCP6.0 temperature')
ax[1,1].set_title('RCP8.5 temperature')
ax[0,0].legend()
```

```python
fig, ax = pl.subplots(figsize=(16,9))
ax.plot(np.arange(1765, 2501),T['rcp26_inc']-T['rcp26_exc'], label='RCP2.6')
ax.plot(np.arange(1765, 2501),T['rcp45_inc']-T['rcp45_exc'], label='RCP4.5')
ax.plot(np.arange(1765, 2501),T['rcp60_inc']-T['rcp60_exc'], label='RCP6.0')
ax.plot(np.arange(1765, 2501),T['rcp85_inc']-T['rcp85_exc'], label='RCP8.5')
ax.legend()
pl.title('Difference between including and excluding CH4 oxidation to CO2 on temperature')
```
